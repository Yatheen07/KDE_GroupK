```python
gdpData = pd.read_csv('/content/GDP.csv')
del gdpData["Unnamed: 0"]
gdpData.head()
```

```python
gdpGrowth = pd.read_csv('/content/GDPGrowth.csv')
del gdpGrowth['Unnamed: 0']
gdpGrowth.head()
```

```python
military = pd.read_csv('/content/MilitaryExpenditure (1).csv')
del military['Unnamed: 0']
military.head()
```

```python
data = []

result = {}
for index,row in gdpData.iterrows():
  temp = {}
  country = row['Entity']
  id = row["ID"]
  year = row["Year"]
  gdp = row["GDP"]
  temp["GDP"] = gdp
  temp["Country"] = country
  temp["Year"] = year
  result[id] = temp

print(result)
```

```python
for index,row in gdpGrowth.iterrows():
  id = row["ID"]
  if id in result:
    temp = result[id]
    data = row["GDPGrowth"]
    temp["GDPGrowth"] = data
    result[id] = temp
print(result["Afghanistan_2018"])
```

```python
for index,row in military.iterrows():
  id = row["ID"]
  if id in result:
    temp = result[id]
    data = row["MilitaryExpenditure"]
    temp["MilitaryExpenditure"] = data
    result[id] = temp
print(result["Afghanistan_2018"])
```

```python
test = pd.DataFrame.from_dict(result)
test.T.to_csv('/content/Economy.csv')
```
